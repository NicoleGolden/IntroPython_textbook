## (2) modify a list

### a) change elements in a list 

```python
# change elements in a list 

motorcycles = ['honda', 'yamaha', 'suzuki']

motorcycles[0] = 'ducati'        # access 1st element and assign a new value

print(motorcycles)
```

### b) append() function - add elements to end of a list 

```python
# append elements to end of a list 

motorcycles = ['honda', 'yamaha', 'suzuki']

motorcycles.append('ducati')    # use append() function to add element in the end
print(motorcycles) 
```

```python
# create an empty list  & add elements later 

motorcycles = []     # create an empty list first 

motorcycles.append('honda')           # add elements 
motorcycles.append('yamaha')
motorcycles.append('ducati')

print(motorcycles)
```

### c) insert() function - insert elements into a list 

```python
# insert elements into a list 

motorcycles = ['honda', 'yamaha', 'suzuki']

motorcycles.insert(1, 'ducati')       # use insert() function, fix 2nd position, and add element

print(motorcycles)
```

### d) "del" statement - remove elements from a list 

```python
# "del" statement - once delete, can no longer access

motorcycles = ['honda', 'yamaha', 'suzuki']

del motorcycles[0]      # use "del" statement, access  1st element, delete it.
print(motorcycles)
```

### e)  pop() function - remove elements from a list

```python
# pop() function - once delete, can  still access

# pop the last element by default 
motorcycles = ['honda', 'yamaha', 'suzuki']

pop_motorcycles = motorcycles.pop()    # pop the last element by default

print(pop_motorcycles)
```

```python
# pop the 1st element

motorcycles = ['honda', 'yamaha', 'suzuki']

pop_motorcycles = motorcycles.pop(0)      # pop the 1st element

print(pop_motorcycles)    

print(motorcycles)            # once poped 1st element, the original list doesn't have it anymore
```

### f) remove() function - remove elements from a list 

```python
# remove() function - Only remove 1st occurrence. To remove all, need a loop

motorcycles = ['honda', 'yamaha', 'suzuki']

motorcycles.remove('honda')      # pop value 'honda'

print(motorcycles)            
```

### b) let the user choose when to quit

```python
prompt = "\nTell me something, and I'll repeat it back to you:"
prompt += "\nEnter 'quit' to end te program."

message = ""                              # set up empty message to store whatever you input 
while message != 'quit':
    message = input(prompt)
    
    if message != 'quit':
        print(message)
```

### c) use a flag

```python
prompt = "\nTell me something, and I'll repeat it back to you:"
prompt += "\nEnter 'quit' to end the program."

active = True           # set "flag" here, as long as it's active, the game keeps going. 
while active: 
    message = input(prompt)
    
    if message == 'quit':
        active = False
    else:
        print(message)
```

### d) use break to exit a loop

## (3) Using while loop with lists and dictionaries 

### a) move items from one list to another

```python
# start with users that need to be verified, and an empty list to hold confirmed users. 
unconfirmed_users = ['alice', 'lee', 'may']
confirmed_users = []

# verify each user until there are no more unconfirmed users. 
# move each verified user into list of confirmed users. 
while unconfirmed_users:
    current_user = unconfirmed_users.pop()
    
    print("Verifying user: " + current_user.title())
    confirmed_users.append(current_user)
    
# display all confirmed users.
print("\nThe following users have been confirmed:")
for confirmed_user in confirmed_users:
    print(confirmed_user.title())
```

### b) remove all instances of specific values from a list 

```python
pets = ['dog', 'cat', 'dig', 'fish', 'cat', 'rabbit', 'cat']
print(pets)

while 'cat' in pets: 
    pets.remove('cat')

print(pets)
```

### c) fill a dictionary with user input

```python
responses = {}

# set a flag to indicate that polling is active. 
polling_active = True

while polling_active:
    
    # prompt for person's name and response
    name = input("\nWhat's your name?")     
    response = input("Which mountain would you like to climb someday?")
    
    # store response in dictionary: 
    responses[name] = response
    
    # find out if anyone else is going to take the poll. 
    repeat = input("Would you like to let another person response(yes/ no) ")
    if repeat == 'no':
        polling_active = False
    
# polling is complete, show results. 
print("\n--- Poll Results ---")
for name, response in responses.items():
    print(name + " would like to climb " + response + ".")
    
```

## (2) passing argument

### a) positional arguments

```python
def describe_pet(animal_type, pet_name): 
    """Display information about a pet."""
    print("\nI have a " + animal_type + ".")
    print("My " + animal_type + "'s name is " + pet_name.title() + ".")

#### Examples to use the function 
#  (1) Order matters in positional arguments (animal_type, pet_name)
describe_pet('hampster', 'harry')

# (2) You can call the function multiple times  
describe_pet('hampster', 'harry')
describe_pet('dog', 'henry')

#  (3) If specify argument (animal_type, pet_name), then order doesn't matter
describe_pet(pet_name =  'harry', animal_type = 'hampster')
```

### b) Default values

```python
# set animal_type = 'dog'

def describe_pet(pet_name, animal_type = 'dog'): 
    """Display information about a pet."""
    print("\nI have a " + animal_type + ".")
    print("My " + animal_type + "'s name is " + pet_name.title() + ".")

#### Example 
# (1) use default 'dog'
describe_pet(pet_name = 'will')

# (2) change animal type to 'cat'
describe_pet(pet_name = 'will', animal_type = 'cat')
```

## (3) Return Values

### a) return a simple value 

```python
# print out full name neatly

def get_formatted_name(first_name, last_name): 
    """Return a full name, neatly formatted."""
    full_name = first_name + " " + last_name
    return full_name.title()

musician = get_formatted_name('jim', 'kelly')
print(musician)
```

### b) make an argument optional

```python
# print out full name neatly: first, middle, last name

def get_formatted_name(first_name, middle_name, last_name): 
    """Return a full name, neatly formatted."""
    full_name = first_name + " " + middle_name + " " + last_name
    return full_name.title()

musician = get_formatted_name('jim', 't.' , 'kelly')
print(musician)
```

```python
# make middle name optional

def get_formatted_name(first_name, last_name, middle_name = ' '):     #  middle_name = ' ' makes it optional
    """Return a full name, neatly formatted."""
    if middle_name: 
        full_name = first_name + " " + middle_name + " " + last_name
    else:
        full_name = first_name + " " + last_name
    return full_name.title()

musician = get_formatted_name('jim', 'kelly', 'T.')
print(musician)
```

### c) return a dictionary 

```python
def build_person(first_name, last_name):
    """Return a dictionary of information about a person."""
    person = {'first': first_name, 'last': last_name}
    return person

musician = build_person('bruce', 'lee')
print(musician)
```

```python
# add 'age' optional

def build_person(first_name, last_name, age=''):
    """Return a dictionary of information about a person."""
    person = {'first': first_name, 'last': last_name}
    if age: 
        person['age'] = age
    return person

musician = build_person('bruce', 'lee', age = 27)
print(musician)
```
